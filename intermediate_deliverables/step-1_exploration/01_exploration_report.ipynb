    2) Définition du périmètre d'étude : critères de sélection des images à récupérer

La fonction de notre projet est de classer les champignons. L'enjeu est de se placer au bon niveau dans la taxonomie des champignons : trop bas et il n'y aura pas assez de photo pour permettre au modèle de s'entrainer, trop haut et les différences physiques des champignons ne seront pas assez marquées pour permettre au modèle de trouver des patterns. Il y a également des contraintes à prendre en compte en termes d'espace de stockage (quelques Go à plusieurs dizaines..) et de temps pour récupérer les photos.

Dans un premier temps nous avons restreint le niveau auquel se placer dans la taxonomie entre famille, genre et espèces.

Les critères de sélection du niveau adéquat entre famille, genre et espèce sont

    a) Nombre de photos par classe : intuitivement, on considère qu'il faut plus de 100 photos pour entraîner le modèle. Nous avons également constaté une assymétrie dans le nombre de champgignons par classe : par exemple certains genres de champignons peuvent avoir plus de 10k photos alors que d'autres moins de 10, ce qui pourrait générer des problèmes ensuite dans la performance du modèle. Il faudrait un nombre de photos à peu près équivalent pour chaque classe.  Dans la création de notre dataset, nous avons fixé une limite maximum de nombre de photos par classe, à 2000 photos, quitte à le réduire ensuite dans la phase de modélisation si besoin. => on se placerait plutôt au niveau genre

    b) ressemblance physique : après quelques analyses rapides de photos, les champignons semblent avoir plus de caractérisques communes au niveau du genre qu'au niveau de la famille. => on se placerait plutôt au niveau genre

    c) intérêt et pertinence du projet : pour un passionné de champignon, il est plus d'intéressant d'obtenir des informations les plus précises et pertinentes possibles => on se placerait plutôt au niveau espèce

En conclusion, nous retenons de déterminer le genre du champignon qui nous semble être un bon compromis. Nous partons dans un premier temps de l'ensemble des genres disponibles contenant au moins 1 photo (maximum 2000). Les genres qui ne contiennent pas assez de photos (moins de 100 [à confirmer] photos) seront supprimés du périmètre

En plus des images, nous avons récupéré les informations suivantes: 
    - genre, famille, ordre du champignon
    - lieu géographique de prise de la photo (indiquer le lieu lors du test du modèle permettrait d'affiner les résultats?)
    - date de prise de l'image : permettra de créer un trainset, test et validation set (images les plus anciennes/récentes)
